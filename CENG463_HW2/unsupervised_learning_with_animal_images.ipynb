## **Introduction**

*In our assignment where we explore unsupervised learning clustering methods using animal images (cat, frog, horse), our objective was to visualize scores (Silhouette and Ground Truth - ARI) by experimenting with various strategies such as feature extraction and dimensionality reduction, including PCA, t-SNE, and NMF, before applying three different clustering methods (K-Means, Agglomerative, DBSCAN).*

### **Creating the Dataset**

*This Python code performs various tasks using the CIFAR-10 dataset. Initially, the CIFAR-10 dataset is loaded using the cifar10.load_data() function, and specific classes are selected. The selected class names and corresponding labels are printed. Subsequently, 500 randomly chosen examples from each selected class are prepared by converting them to grayscale, adding the "Gray Scale" label, normalizing, and flattening the images. The dataset shape and the dimensions of grayscale images are then printed. Finally, the dataset is standardized using StandardScaler, centering the features around the mean and scaling by the standard deviation. The standardized dataset is stored for potential further analysis or learning processes in the "flattened_images_scaled" variable.*

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.preprocessing import StandardScaler
from skimage.color import rgb2gray

# Load CIFAR-10 dataset
(x_train, y_train), _ = cifar10.load_data()

# Define the class names in CIFAR-10
cifar10_class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Select three classes (e.g., cat=3, frog=6, horse=7)
classes_to_include = [3, 6, 7]
class_names = [cifar10_class_names[i] for i in classes_to_include]

# Print class names and their corresponding labels in the dataset
print("Selected class names:", class_names)

# Select and preprocess images
selected_indices = np.hstack([np.random.choice(np.where(y_train.flatten() == cls)[0], 1000) for cls in classes_to_include])
selected_images = x_train[selected_indices]
selected_labels = y_train[selected_indices].flatten()

# Convert images to grayscale
gray_images = rgb2gray(selected_images) / 255.0

# Flatten images
flattened_images = gray_images.reshape(len(gray_images), -1)

# Print dataset characteristics
print("Dataset shape:", flattened_images.shape)
print("Image size:", gray_images.shape[1:3])

# Standard Scaler
scaler = StandardScaler()
flattened_images_scaled = scaler.fit_transform(flattened_images)
```

### **Comparative Analysis of Clustering Algorithms on PCA**

*This Python code performs dimensionality reduction using PCA (Principal Component Analysis) with two components and subsequently applies different clustering algorithms, including KMeans, Agglomerative Clustering, and DBSCAN. Initially, the data is reduced to two components using PCA with the n_components parameter set to 2. Following this, KMeans, Agglomerative Clustering, and DBSCAN algorithms are applied to the reduced dataset. Metrics evaluating clustering performance, such as Silhouette and Adjusted Rand Index (ARI), are calculated for each algorithm. A special procedure is applied to exclude noise when using DBSCAN. Finally, the clustering results are visualized through plots, presenting the performance of each algorithm with Silhouette and ARI scores in the titles. This analysis is designed to compare the performance of clustering algorithms applied to the dimensionally reduced version of the dataset.*

```python
# Dimensionality reduction with PCA
pca = PCA(n_components=2)
pca_images = pca.fit_transform(flattened_images_scaled)

# Clustering algorithms
algorithms = {
    'KMeans': KMeans(n_clusters=len(classes_to_include)),
    'Agglomerative': AgglomerativeClustering(n_clusters=len(classes_to_include)),
    'DBSCAN': DBSCAN(eps=0.5)
}

# Fit models and calculate metrics
scores = {}
for name, algorithm in algorithms.items():
    algorithm.fit(pca_images)
    labels = algorithm.labels_

    # Exclude noise for DBSCAN
    if name == 'DBSCAN' and np.min(labels) < 0:
        valid_indices = labels != -1
        labels_filtered = labels[valid_indices]
        pca_subset = pca_images[valid_indices]
        selected_subset = selected_labels[valid_indices]
    else:
        labels_filtered = labels
        pca_subset = pca_images
        selected_subset = selected_labels

    # Calculate scores
    silhouette = silhouette_score(pca_subset, labels_filtered) if len(np.unique(labels_filtered)) > 1 else 'N/A'
    ari = adjusted_rand_score(selected_subset, labels_filtered)
    scores[name] = (silhouette, ari)

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
for ax, (name, algorithm) in zip(axes, algorithms.items()):
    labels = algorithm.labels_

    # Exclude noise for DBSCAN
    if name == 'DBSCAN' and np.min(labels) < 0:
        valid_indices = labels != -1
        labels_valid = labels[valid_indices]
        pca_subset = pca_images[valid_indices]
        ax.scatter(pca_subset[:, 0], pca_subset[:, 1], c=labels_valid, cmap='viridis', marker='o')
    else:
        ax.scatter(pca_images[:, 0], pca_images[:, 1], c=labels, cmap='viridis', marker='o')

    ax.set_title(f'{name} Clustering\nSilhouette: {scores[name][0]}\nARI: {scores[name][1]}')
    ax.set_xlabel('PCA Feature 1')
    ax.set_ylabel('PCA Feature 2')

plt.tight_layout()
plt.show()
```

### **Comparative Analysis of Clustering Algorithms on t-SNE**

*This Python code employs t-SNE (t-distributed Stochastic Neighbor Embedding) for dimensionality reduction to two components and evaluates the performance of clustering algorithms on the resulting reduced CIFAR-10 dataset. The clustering algorithms, including KMeans, Agglomerative Clustering, and DBSCAN, are applied to the t-SNE-transformed dataset. Metrics such as Silhouette and Adjusted Rand Index (ARI) are computed to assess the quality of clustering for each algorithm. A special procedure is implemented to handle noise exclusion in the case of DBSCAN. The clustering results are visualized through plots, depicting the distribution of clusters in the two-dimensional t-SNE space, and the performance metrics are displayed in the titles. This analysis aims to compare the effectiveness of clustering algorithms when applied to the dataset reduced using t-SNE.*

```python
from sklearn.manifold import TSNE


# t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
tsne_images = tsne.fit_transform(flattened_images_scaled)

# Clustering algorithms
algorithms = {
    'KMeans': KMeans(n_clusters=len(classes_to_include)),
    'Agglomerative': AgglomerativeClustering(n_clusters=len(classes_to_include)),
    'DBSCAN': DBSCAN(eps=0.5)
}

# Fit models and calculate metrics
scores = {}
for name, algorithm in algorithms.items():
    algorithm.fit(tsne_images)
    labels = algorithm.labels_

    # Exclude noise for DBSCAN
    if name == 'DBSCAN':
        valid_indices = labels != -1
        if np.any(valid_indices):  # Check if there are non-outlier points
            labels_valid = labels[valid_indices]
            tsne_valid = tsne_images[valid_indices]
            selected_valid = selected_labels[valid_indices]
            silhouette = silhouette_score(tsne_valid, labels_valid) if len(np.unique(labels_valid)) > 1 else 'N/A'
            ari = adjusted_rand_score(selected_valid, labels_valid)
    else:
        silhouette = silhouette_score(tsne_images, labels) if len(np.unique(labels)) > 1 else 'N/A'
        ari = adjusted_rand_score(selected_labels, labels)

    scores[name] = (silhouette, ari)

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
for ax, (name, algorithm) in zip(axes, algorithms.items()):
    labels = algorithm.labels_

    # Exclude noise for DBSCAN
    if name == 'DBSCAN':
        valid_indices = labels != -1
        labels_valid = labels[valid_indices]
        tsne_valid = tsne_images[valid_indices]
        ax.scatter(tsne_valid[:, 0], tsne_valid[:, 1], c=labels_valid, cmap='viridis', marker='o')
    else:
        ax.scatter(tsne_images[:, 0], tsne_images[:, 1], c=labels, cmap='viridis', marker='o')

    ax.set_title(f'{name} Clustering\nSilhouette: {scores[name][0]}\nARI: {scores[name][1]}')
    ax.set_xlabel('t-SNE Feature 1')
    ax.set_ylabel('t-SNE Feature 2')

plt.tight_layout()
plt.show()
```

### **Exploring Dimensionality Reduction with NMF**

*This Python code leverages Non-Negative Matrix Factorization (NMF) for dimensionality reduction to two components and evaluates the performance of clustering algorithms on the resulting reduced CIFAR-10 dataset. The NMF algorithm is applied to flatten images, and KMeans, Agglomerative Clustering, and DBSCAN are subsequently employed for clustering. The metrics, including Silhouette and Adjusted Rand Index (ARI), are computed to measure the quality of clustering for each algorithm. A specific handling for noise exclusion is implemented for DBSCAN. The clustering results are visualized through plots, displaying the distribution of clusters in the two-dimensional NMF space, and the performance metrics are presented in the titles. This analysis aims to investigate the effectiveness of clustering algorithms on the CIFAR-10 dataset after dimensionality reduction using NMF.*

```python
from sklearn.decomposition import NMF

# NMF for dimensionality reduction
nmf = NMF(n_components=2, init='random', random_state=42, max_iter=1000)
nmf_images = nmf.fit_transform(flattened_images)

# Clustering algorithms
algorithms = {
    'KMeans': KMeans(n_clusters=len(classes_to_include)),
    'Agglomerative': AgglomerativeClustering(n_clusters=len(classes_to_include)),
    'DBSCAN': DBSCAN(eps=0.5)
}

# Fit models and calculate metrics
scores = {}
for name, algorithm in algorithms.items():
    algorithm.fit(nmf_images)
    labels = algorithm.labels_

    # Exclude noise for DBSCAN
    if name == 'DBSCAN' and np.min(labels) < 0:
        valid_indices = labels != -1
        labels_valid = labels[valid_indices]
        nmf_valid = nmf_images[valid_indices]
        selected_valid = selected_labels[valid_indices]
        silhouette = silhouette_score(nmf_valid, labels_valid) if len(np.unique(labels_valid)) > 1 else 'N/A'
        ari = adjusted_rand_score(selected_valid, labels_valid)
    else:
        silhouette = silhouette_score(nmf_images, labels) if len(np.unique(labels)) > 1 else 'N/A'
        ari = adjusted_rand_score(selected_labels, labels)

    scores[name] = (silhouette, ari)

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
for ax, (name, algorithm) in zip(axes, algorithms.items()):
    labels = algorithm.labels_

    # Exclude noise for DBSCAN
    if name == 'DBSCAN' and np.min(labels) < 0:
        valid_indices = labels != -1
        labels_valid = labels[valid_indices]
        nmf_valid = nmf_images[valid_indices]
        ax.scatter(nmf_valid[:, 0], nmf_valid[:, 1], c=labels_valid, cmap='viridis', marker='o')
    else:
        ax.scatter(nmf_images[:, 0], nmf_images[:, 1], c=labels, cmap='viridis', marker='o')

    ax.set_title(f'{name} Clustering\nSilhouette: {scores[name][0]}\nARI: {scores[name][1]}')
    ax.set_xlabel('NMF Feature 1')
    ax.set_ylabel('NMF Feature 2')

plt.tight_layout()
plt.show()
```
